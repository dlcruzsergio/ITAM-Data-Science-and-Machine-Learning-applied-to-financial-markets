```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
#metadata = pd.read_csv("../files/lec02/world-bank/metadata.csv", index_col=0).iloc[:, :-1]
gdp = pd.read_csv("../files/lec02/world-bank/gdp_per_capita.csv", skiprows=3).drop("Unnamed: 64", axis=1)
metadata = pd.read_csv("../files/lec02/world-bank/metadata.csv").drop("Unnamed: 5", axis=1)
children = pd.read_csv("../files/lec02/world-bank/children_out.csv", skiprows=3).drop("Unnamed: 64", axis=1)

print("******************************************")
print("*** GENERAL INFORMATION ABOUT METADATA ***")
print("******************************************", end="\n\n")
print(f"Columns ({len(metadata.columns)}):\n{metadata.columns}", end="\n\n")
print(f"Countries ({len(metadata['Country Code'].unique())}):\n{metadata['Country Code'].unique()}", end="\n\n")
```

```python
metadata.head(2)
```

```python
gdp.head(2)
```

```python
metadata_gdp = metadata.set_index("Country Code").join(gdp.set_index("Country Code"), how="left")
metadata_gdp_by_income = metadata_gdp.set_index("IncomeGroup").iloc[:, 7:].groupby("IncomeGroup")
metadata_gdp_by_income.sum()
```

```python
y = metadata_gdp_by_income.sum().T["High income"]

plt.figure(1); plt.title('High income - Johnson SU')
sns.distplot(y, kde=False, fit=stats.johnsonsu)

plt.figure(2); plt.title('High income - Normal')
sns.distplot(y, kde=False, fit=stats.norm)

plt.figure(3); plt.title('High income - Log Normal')
sns.distplot(y, kde=False, fit=stats.lognorm)
```

```python
y = metadata_gdp_by_income.sum().T["Low income"]

plt.figure(1); plt.title('Low income - Johnson SU')
sns.distplot(y, kde=False, fit=stats.johnsonsu)

plt.figure(2); plt.title('Low income - Normal')
sns.distplot(y, kde=False, fit=stats.norm)

plt.figure(3); plt.title('Low income - Log Normal')
sns.distplot(y, kde=False, fit=stats.lognorm)
```

```python
y = metadata_gdp_by_income.sum().T["Lower middle income"]

plt.figure(1); plt.title('Lower middle income - Johnson SU')
sns.distplot(y, kde=False, fit=stats.johnsonsu)

plt.figure(2); plt.title('Lower middle income - Normal')
sns.distplot(y, kde=False, fit=stats.norm)

plt.figure(3); plt.title('Lower middle income - Log Normal')
sns.distplot(y, kde=False, fit=stats.lognorm)
```

```python
y = metadata_gdp_by_income.sum().T["Lower middle income"]

plt.figure(1); plt.title('Upper middle income - Johnson SU')
sns.distplot(y, kde=False, fit=stats.johnsonsu)

plt.figure(2); plt.title('Upper middle income - Normal')
sns.distplot(y, kde=False, fit=stats.norm)

plt.figure(3); plt.title('Upper middle income - Log Normal')
sns.distplot(y, kde=False, fit=stats.lognorm)
```

```python
fig, ax = plt.subplots(2,2, figsize=(18,10))
fig.suptitle('IncomeGroup sum', fontsize=24)

metadata_gdp_by_income.sum().T.plot(ax=ax[0,0]);
metadata_gdp_by_income.sum().drop("High income").T.plot(ax=ax[0,1]);

pd.DataFrame(data=[
    metadata_gdp_by_income.sum().T["High income"],
    metadata_gdp_by_income.sum().T["Low income"],
    metadata_gdp_by_income.sum().T["Lower middle income"],
    metadata_gdp_by_income.sum().T["Upper middle income"]]).T.plot(ax=ax[1,0], kind='box');
pd.DataFrame(data=[
    metadata_gdp_by_income.sum().T["Low income"],
    metadata_gdp_by_income.sum().T["Lower middle income"],
    metadata_gdp_by_income.sum().T["Upper middle income"]]).T.plot(ax=ax[1,1], kind='box');
```
